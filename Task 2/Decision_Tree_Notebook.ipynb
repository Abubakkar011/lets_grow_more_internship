# <center> DECISION TREE CLASSIFICATION </center>

## Importing Dependencies

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import confusion_matrix
from seaborn import heatmap
from sklearn.datasets import load_iris
```

## Reading CSV file

```python
df=pd.read_csv('iris.csv')
df
```

## Splitting Train and Test Data

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['target']),df['target'])

model=DecisionTreeClassifier()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

## Making Predictions of x_test

```python
y_pred=model.predict(x_test)
```

## Comparing Actual data vs. Predicted data

```python
actual_x_predicted=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
actual_x_predicted.head()
```

## Saving the comparision file

```python
actual_x_predicted.to_csv('Actual vs Predicted.csv')
```

## Applying Confusion Matrix to fetch the error count

```python
heatmap(confusion_matrix(y_test,y_pred),annot=True)
```

## Plotting the Decision Tree

```python
plt.figure(figsize=(10,10))
plot_tree(model,filled=True,fontsize=7)
plt.savefig('decision_tree.png')
```
